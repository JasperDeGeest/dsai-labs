## Exercise 1 - Cholesterol 

A health survey between 1988 and 1994 indicated that the mean cholesterol value for women between the age of 20 and 29 was 183 mg/dl, with a standard deviation equal to 36. We now take a random sample of 81 women.

1. Plot the probability distribution of the sample mean $\overline{x}$.
2. What is the probability of the sample mean being $\overline{x} < 185$? (Answer ‚âà 69,1%)
3. What is the probability of $175 < \overline{x} < 185$ (Answer ‚âà 66,9%)
4. What is the probability of $\overline{x} > 190$ (Answer ‚âà 4,0%)

```python
# Parameters
mu = 183                # population mean
sigma = 36              # population std deviation
n = 81                  # sample size
se = sigma / np.sqrt(n) # standard error = 36 / ‚àö81 = 4
```

```python
# X range for plot
x = np.linspace(mu - 4 * se, mu + 4 * se, num=1000)
pdf = stats.norm.pdf(x, mu, se)

# 1. Plot probability distribution of the sample mean
plt.figure(figsize=(5, 3))
plt.plot(x, pdf, label='Sampling distribution of $\overline{x}$', color='blue')
plt.title('Sampling Distribution of the Sample Mean')
plt.xlabel('Sample Mean $\overline{x}$')
plt.ylabel('Probability Density')
plt.legend()

# 2. P(xÃÑ < 185)
p1 = stats.norm.cdf(185, mu, se)

# 3. P(175 < xÃÑ < 185)
p2 = stats.norm.cdf(185, mu, se) - stats.norm.cdf(175, mu, se)

# 4. P(xÃÑ > 190)
p3 = 1 - stats.norm.cdf(190, mu, se)

# Show the plot
plt.show()

# Print probabilities
print(f"2. P(ùë•ÃÑ < 185) ‚âà {p1:.1%}")
print(f"3. P(175 < ùë•ÃÑ < 185) ‚âà {p2:.1%}")
print(f"4. P(ùë•ÃÑ > 190) ‚âà {p3:.1%}")
```

## Exercise 2

A random sample of 64 is drawn from a population with unknown distribution. However, the expectation and standard deviation of the population are known: $\mu = 20$ and $\sigma = 16$.

1. Plot the probability distribution of the sample mean.
2. Calculate the z-score for $\overline{x_1} = 15.5$  
3. Calculate the z-score for $\overline{x_2} = 23$ 
4. Determine the probability that 16 < $\overline{x}$ < 22 (Answer ‚âà 81.9%)

```python
# Parameters
mu = 20               # population mean
sigma = 16             # population std deviation
n = 64                 # sample size
se = sigma / np.sqrt(n) 
# X range for plot
x = np.linspace(mu - 4 * se, mu + 4 * se, num=1000)
pdf = stats.norm.pdf(x, mu, se)

# 1. Plot probability distribution of the sample mean
plt.figure(figsize=(5, 3))
plt.plot(x, pdf, color='blue')

print("2. z-score for x1: ",(15.5 - mu) / se)
print("3. z-score for x2: ",(23 - mu) / se)

print("4. P(16 < x < 22): %.3f" % (stats.norm.cdf(22, mu, se)-stats.norm.cdf(16, mu, se)))

```
